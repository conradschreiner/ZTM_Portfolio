```python
# alternate method
df_wv = pd.DataFrame(fifa_df, columns=['Name', 'Wage', 'Value'])
df_wv.head()
```

```python
df_wv.info()
```

```python
#converting wage and value from string to int
def value_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000.0
    if 'B' in x:
        return float(x.replace('B', '')) * 1000000000
    return 0.0

wage = df_wv['Wage'].replace('[\€,]', '', regex=True).apply(value_to_float)
value = df_wv['Value'].replace('[\€,]', '', regex=True).apply(value_to_float)
df_wv['Wage'] = wage
df_wv['Value'] = value
df_wv['Value'] = df_wv['Value'].apply(value_to_float)
df_wv['Wage'] = df_wv['Wage'].apply(value_to_float)
```

```python
df_wv['difference'] = df_wv['Value'] - df_wv['Wage']
difference_sorted = df_wv.sort_values('difference', ascending=False)
```

```python
difference_sorted.head()
```

```python
import seaborn as sns
from bokeh.plotting import figure,show
from bokeh.models import HoverTool
```

```python
sns.set()
wv_graph = sns.scatterplot(x='Wage', y='Value', data=df_wv)
display(wv_graph)
```

```python
TOOLTIPS = HoverTool(tooltips=[
    ("index", "$index"),
    ("(Wage, Value)", "(@Wage, @Value)"),
    ("Name", "@Name")]
)
p = figure(title="FIFA Wage-Value Stats 2019", x_axis_label='Wage', y_axis_label='y',
          plot_width=700, plot_height=700, tools=[TOOLTIPS])
p.circle('Wage', 'Value', size=10, source=df_wv)
show(p)
```
